# **1. Mean**

Average value of all values formula Mean =  Σx/n

1.  Σx = Sum of all observations
2. n = Numbers of observation

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset")
print(df)

# =====================================
# Mean of All Numerical Columns
# =====================================

print("\nMean of Numerical Columns")
print(df.mean(numeric_only=True))

# =====================================
# Mean of Individual Columns
# =====================================

print("\nAverage Age:", df["Age"].mean())
print("Average Salary:", df["Salary"].mean())
print("Average Experience:", df["Experience"].mean())
```

# **2. Mode**

the Mode is the value that appears **most** frequently

```python
import pandas as pd
import numpy as np


data = {
    "Employee": [
        "A", "B", "C", "D", "E",
        "F", "G", "H", "I", "J"
    ],
    "Age": [
        22, 25, 28, 30, 35,
        40, 32, 29, 45, 38
    ],
    "Salary": [
        25000, 32000, 40000, 42000, 50000,
        62000, 47000, 43000, 75000, 68000
    ],
    "Experience": [
        1, 2, 4, 5, 8,
        12, 6, 5, 18, 15
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset
print("Employee Dataset")
print(df)

# ============================================
# Calculate Mode
# ============================================

print("\nMode of Each Column")
print(df.mode())

# ============================================
# Mode of Individual Columns
# ============================================

print("\nMode of Age")
print(df["Age"].mode())

print("\nMode of Salary")
print(df["Salary"].mode())

print("\nMode of Experience")
print(df["Experience"].mode())

# ============================================
# First Mode Value (if multiple modes exist)
# ============================================

print("\nFirst Mode of Experience:", df["Experience"].mode()[0])
```

# **2. Measure of Dispersion**

Measure of dispersion describe the variability or spread of data in a dataset

**1. Range**

The simplest measure of dispersion

**Formula**

Range = Maximum Value - Minimum Value

```python
import numpy as np

data = [12,18,20,25,30]

print("Range =", max(data) - min(data))
```

**2. Interquartile Range**

IQR measures the spread of the middle 50% of the data.
It ignores extreme values

**Formula**

IQR = Q3 - Q1

Where:
 1. Q1 = First Quartile(25%)
 2. Q3 = Third Quartile(75%)

```python
import numpy as np

data = [7, 8, 5, 6, 3, 4, 9, 12, 15, 18]

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("Interquartile Range (IQR):", IQR)
```

# **5. Variance**

Variance measures the average squared deviation from the mean.
A small variance means data points are close to the mean, while a large variance means they are widely spread

**Formula**

var(X) = E(X2) - [E(X)]2

There are two types:

**Population Variance**

σ2 = Σ(x-u)

s2 = Σ(x- x bar)2/n-1

```python
import numpy as np

data = [2, 4, 6, 8]
print("Population Varaince =",np.var(data))

print("Sample Variance =",np.var(data,ddof=1))
```

**6. Standard Deviation**

Standard Deviation is the square root of variance.
It is the most widely used measure of dispersion because it is expressed in the same units as the original data.

**Formula**

**Population Standard Deviation**

σ = σ2

**Sample Standard Deviation**

s = s2

```python
import numpy as np

data = [2, 4, 6, 8]
print("Population SD =",np.std(data))

print("Sample SD =",np.std(data,ddof=1))
```

3. **Measure of Distribution**

Measures of distribution describe how data values are spread and shaped within a dataset.

**Types of Measures of Distribution**

**1. Frequency Distribution**

A frequency distribution shows how many times each value or group of values appers in a dataset


```python
import pandas as pd

marks = [45, 52, 65, 68, 72, 75, 80, 85, 60, 62, 55]

df = pd.DataFrame({'Marks': marks})

print(df["Marks"].value_counts().sort_index())
```

**Skewness**

Skewness measure the symmetry of data

1. Positive - Right-Skewed
2. Negative - Left-Skewed
3. Zero - Symmetric

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew

# ----------------------------
# Create Sample Dataset
# ----------------------------

np.random.seed(42)

# Normally distributed data
data = np.random.normal(loc=50, scale=10, size=500)

df = pd.DataFrame({
    "Marks": data
})


# ----------------------------
# Calculate Skewness
# ----------------------------

skew_value = skew(df["Marks"])

print("\nSkewness Value =", round(skew_value, 4))

# ----------------------------
# Interpretation
# ----------------------------

if skew_value > 0:
    print("Interpretation : Positively (Right) Skewed Distribution")
elif skew_value < 0:
    print("Interpretation : Negatively (Left) Skewed Distribution")
else:
    print("Interpretation : Perfectly Symmetric Distribution")

# ----------------------------
# Histogram
# ----------------------------

plt.figure(figsize=(8,5))

plt.hist(df["Marks"],
         bins=20,
         edgecolor="black")

plt.title("Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()

# ----------------------------
# Density Plot
# ----------------------------

plt.figure(figsize=(8,5))

df["Marks"].plot(kind="density")

plt.title("Density Plot")
plt.xlabel("Marks")

plt.show()

# ----------------------------
# Box Plot
# ----------------------------

plt.figure(figsize=(6,4))

plt.boxplot(df["Marks"])

plt.title("Box Plot")

plt.show()
```

**Kurtosis**

Kurtosis measure the heaviness of the tails of the distribution

1. Positive  Sharp tails
2. Negative - Flat and light tails
3. Approx zero -> Normal

```python
# ===========================================
# Kurtosis Complete Practical Example
# ===========================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Set random seed for reproducibility
np.random.seed(42)

# -------------------------------------------------
# Generate Different Types of Distributions
# -------------------------------------------------

# 1. Normal Distribution (Mesokurtic)
normal_data = np.random.normal(loc=50, scale=10, size=1000)

# 2. Heavy-Tailed Distribution (Leptokurtic)
lepto_data = np.random.laplace(loc=50, scale=8, size=1000)

# 3. Uniform Distribution (Platykurtic)
platy_data = np.random.uniform(low=20, high=80, size=1000)

# -------------------------------------------------
# Create DataFrame
# -------------------------------------------------

df = pd.DataFrame({
    "Normal": normal_data,
    "Leptokurtic": lepto_data,
    "Platykurtic": platy_data
})

print("First Five Rows")
print(df.head())

# -------------------------------------------------
# Calculate Kurtosis
# fisher=True
# Normal Distribution = 0
# -------------------------------------------------

print("\n========== Kurtosis Values ==========\n")

for column in df.columns:
    value = kurtosis(df[column], fisher=True)

    print(f"{column} : {value:.4f}")

    if value > 0:
        print(" Interpretation : Leptokurtic (Sharp Peak, Heavy Tails)\n")

    elif value < 0:
        print(" Interpretation : Platykurtic (Flat Peak, Light Tails)\n")

    else:
        print(" Interpretation : Mesokurtic (Normal Distribution)\n")

# -------------------------------------------------
# Plot Histograms
# -------------------------------------------------

plt.figure(figsize=(18,5))

plt.subplot(1,3,1)
plt.hist(df["Normal"], bins=30)
plt.title("Normal Distribution\n(Mesokurtic)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(1,3,2)
plt.hist(df["Leptokurtic"], bins=30)
plt.title("Leptokurtic Distribution")
plt.xlabel("Value")

plt.subplot(1,3,3)
plt.hist(df["Platykurtic"], bins=30)
plt.title("Platykurtic Distribution")
plt.xlabel("Value")

plt.tight_layout()
plt.show()

# -------------------------------------------------
# Density Plots
# -------------------------------------------------

plt.figure(figsize=(10,6))

df["Normal"].plot(kind="density", label="Normal")
df["Leptokurtic"].plot(kind="density", label="Leptokurtic")
df["Platykurtic"].plot(kind="density", label="Platykurtic")

plt.title("Density Plot")
plt.legend()
plt.show()

# -------------------------------------------------
# Box Plot
# -------------------------------------------------

plt.figure(figsize=(8,6))

plt.boxplot(
    [df["Normal"], df["Leptokurtic"], df["Platykurtic"]],
    labels=["Normal","Leptokurtic","Platykurtic"]
)

plt.title("Box Plot Comparison")
plt.ylabel("Values")
plt.show()

# -------------------------------------------------
# Describe Dataset
# -------------------------------------------------

print("\n========== Summary Statistics ==========\n")
print(df.describe())
```

**Quartile**

Quartiles are statistical values that divide a sorted dataset into four equals parts

**Quartiles**         **Percentage of data**         **description**

* Q1                       25%
* Q2                       50%
* Q3                       75%
* Q4                       100%


```python
import pandas as pd
data = {
    "Employee": ["A","B","C","D","E","F","G","H","I","J"],
    "Salary": [25000,32000,40000,42000,50000,
               62000,47000,43000,75000,68000]
}

df = pd.DataFrame(data)

print(df)

Q1 = df["Salary"].quantile(0.25)
Q2 = df["Salary"].quantile(0.50)
Q3 = df["Salary"].quantile(0.75)

print("\nFirst Quartile (Q1):", Q1)
print("Second Quartile (Median/Q2):", Q2)
print("Third Quartile (Q3):", Q3)
```
